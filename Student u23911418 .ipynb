```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('Downloads/u23911418.Compiled Data Set - Sheet1.csv')

# Data Preprocessing
# Check for missing values and handle them
df.dropna(subset=['Nickname'], inplace=True)

# Encode categorical variables: 'Winner' and 'Level'
label_encoder_winner = LabelEncoder()
label_encoder_level = LabelEncoder()
df['Winner_encoded'] = label_encoder_winner.fit_transform(df['Winner'])
df['Level_encoded'] = label_encoder_level.fit_transform(df['Level'])

# Drop unnecessary columns
df_cleaned = df.drop(columns=['Nickname', 'Winner', 'Level'])

# Exploratory Data Analysis (EDA)
# Summary Statistics
print(df_cleaned.describe())

# Histograms of numeric columns
df_cleaned[['Defender Score', 'Attacker Score', 'Time (sec)']].hist(bins=20, figsize=(10, 5))
plt.show()

# Boxplots of scores and time
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
sns.boxplot(y=df_cleaned['Defender Score'])
plt.title('Defender Score')
plt.subplot(1, 3, 2)
sns.boxplot(y=df_cleaned['Attacker Score'])
plt.title('Attacker Score')
plt.subplot(1, 3, 3)
sns.boxplot(y=df_cleaned['Time (sec)'])
plt.title('Time (sec)')
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_cleaned.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
